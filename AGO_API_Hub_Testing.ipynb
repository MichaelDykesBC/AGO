```python
import os
from arcgis.gis import GIS
from arcgishub import hub
```

### You don't need GIS to connect, you can use Hub

```python
AGO_URL = "https://bcgov03.maps.arcgis.com"
AGO_USERNAME = os.environ["GEOHUB_USERNAME"]
AGO_PASSWORD = os.environ["GEOHUB_PASSWORD"]

myHub = hub.Hub(AGO_URL,username=AGO_USERNAME,password=AGO_PASSWORD)
```

### Hub Content Hierarchy: Initiatives -> Sites -> Pages

```python
for row in myHub.initiatives.search():
    print(row)
```

### Hub Content Hierarchy: Initiatives -> Sites -> Pages

```python
hub_initiative_itemid = "d8ebd2ce1a5e4bca88f6160a1a69362f"

myInitiative = myHub.initiatives.get(hub_initiative_itemid)
mySite = myInitiative.site_id

site1 = myHub.sites.get(mySite)
print(site1)
print(site1.url)
```

### Layout -> Section -> Row -> Card

```python
mySite_layout = site1.layout

print(len(mySite_layout.sections))
for row in mySite_layout.sections:
    print("==================================")
    for k, v in row.items():
        print(f'{k} - {v}')
```

### Get the custom HTML data from Text Card (aka Markdown Card)

```python
bier_cards = mySite_layout.sections[2]['rows'][0]['cards'][0]['component']

bier_cards_html = bier_cards['settings']['markdown']

print(bier_cards_html)
```

### Parse HTML using BeautifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(bier_cards_html, 'html.parser')

print(soup)
```

### Write HTML code to file

```python
html_doc = r'C:\Users\Mike\Desktop\hub_demo.html'

with open(html_doc, "w", encoding='utf-8') as file:
    file.write(str(soup))
```
